## Checking the sum of deposits and withdraw

```python
#new
depositSum = 0
withdrawSum = 0

for i,j in zip(df.DEPOSITS, df.WITHDRAWALS):
  if pd.notna(i):
    depositSum += i
  if pd.notna(j):
    withdrawSum += j


print(depositSum)
print(withdrawSum)
```

```python
n = df.shape[0]
```

```python
for i in range(len(df)):
    if pd.isna(df.DATE[i]) or df.DATE[i] == 'nan':  # Check if DATE is missing
        datVal = df.DATE[i - 1]
        val = str(df.PARTICULARS[i - 1])
        #print(datVal)
        #print(val)

        j = i
        while j < len(df) and (pd.isna(df.DATE[j]) or df.DATE[j] == 'nan'):
            parVal = str(df.PARTICULARS[j])
            val = val + parVal
            #print(val)
            j += 1
        if j < len(df):
            df.PARTICULARS[i-1] = val

```

## checking if there is any value where the date in NULL but the deposit or withdraw field is filled

```python
isNull = False
for i in range(n):
  if pd.isna(df.DATE[i]) and ( pd.notna(df.DEPOSITS[i]) or pd.notna(df.WITHDRAWALS[i]) or pd.notna(df.BALANCE[i]) ):
    print(i+2)
    isNull = True

if not isNull:
  print("No Null Values")
```

Checking for the NULL DATE values

```python
nanCount = 0
for i in df.DATE:
  if pd.isna(i):
    nanCount += 1
print(nanCount)
```

Dropping the null date values

```python
df = df.dropna(subset=['DATE'])
```

Checking again now

```python
nanCount = 0
for i in df.DATE:
  if pd.isna(i):
    nanCount += 1
print(nanCount)
```
