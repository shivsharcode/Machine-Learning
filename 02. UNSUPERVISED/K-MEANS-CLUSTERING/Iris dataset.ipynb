## Step 1: import dataset

```python
from sklearn.datasets import load_iris

import pandas as pd
```

```python
# load dataset
iris = load_iris()

print(type(iris))

for key in iris:
    print(key)
```

## Data

```python
print(iris.data.shape)
```

150 samples (rows) \
4 features (columns) <i> viz. Sepal length, Sepal width, Petal len, Petal width

```python
print(iris.data)
```

## Target
Target values for the data, representing the class (species)

```python
print(iris.target)
```

see there are just 3 classs = 0, 1, 2 for the three species of iris

## Target names
